## ------------------------------------------DATE FORMATTING----------------------------------------

### **Let's format the Date Column for the questions to be answered without hassle**


#### 1.Drop the Duplicates in Date column if any

```python
duplicated_rows = df[df.duplicated(['Date'])]
df = df.drop(duplicated_rows.index)
```

#### 2. Format the Date Column to '%d-%m-%Y' format and extract Year, Month, Day, Day of Week and add them as columns of Dataframe.

```python
def p_weekday(x):
    return datetime.datetime.strptime(x, '%d-%m-%Y').weekday()
def p_day(x):
    return datetime.datetime.strptime(x, '%d-%m-%Y').day
def p_month(x):
    return datetime.datetime.strptime(x, '%d-%m-%Y').month
def p_year(x):
    return datetime.datetime.strptime(x, '%d-%m-%Y').year
```

```python
df['Year'], df['Month'],df['Day'],df['Day_of_week'] = df['Date'].apply(lambda x: p_year(x)),df['Date'].apply(lambda x: p_month(x)),df['Date'].apply(lambda x: p_day(x)),df['Date'].apply(lambda x: p_weekday(x))
```

```python
df
```

#### 3.Set Date column as Index of DataFrame

```python
df.set_index('Date', inplace=True)
```
